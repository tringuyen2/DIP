# 	Viết hàm: Tính toán và hiển thị lược đồ ảnh: đầu vào là một ảnh, và số bin ; đầu ra là lược đồ (Lena)



```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# Hàm vẽ histogram của ảnh, đầu vào có 2 tham số là ảnh và số bin, đầu ra là lược đồ
def display_hist(img, bin):
  c = 0
  if np.array(img.shape).shape == (3,):
    w, h, c = img.shape
  else:
    w, h = img.shape

  t = 256//bin

  if (c == 0):
    hist = np.zeros((bin), np.uint32)
    for i in range(w):
      for j in range(h):
        hist[img[i][j]//t] += 1
    plt.plot(hist, color='b')
    plt.show()
    return hist
  else:
    hist_r = np.zeros((bin), np.uint32)
    hist_g = np.zeros((bin), np.uint32)
    hist_b = np.zeros((bin), np.uint32)
    for i in range(w):
      for j in range(h):
        hist_b[img[i][j][0]//t] += 1
        hist_g[img[i][j][1]//t] += 1
        hist_r[img[i][j][2]//t] += 1
      
    plt.plot(hist_b, color='b')
    plt.plot(hist_r, color='r')
    plt.plot(hist_g, color='g')
    plt.show()
    return hist_b, hist_g, hist_r


# Câu lệnh thực thi
display_hist(img, 256)
```

# Viết hàm: Thay đổi ảnh sử dụng co giãn lược đồ và cân bằng lược đồ, hiển thị và ghi lại ảnh sau khi đã thực hiện phép biến đổi (Contrast.gif)

```python
# Hàm cân bằng histogram

def equal_hist(hist):
    cumulator = np.zeros_like(hist, np.float64)
    for i in range(len(cumulator)):
        cumulator[i] = hist[:i].sum()
    print(cumulator)
    new_hist = (cumulator - cumulator.min())/(cumulator.max() - cumulator.min()) * 255
    new_hist = np.uint32(new_hist)
    return new_hist
```

```python
#hist = hist.ravel()
new_hist = equal_hist(hist)

h, w = img.shape[:2]
for i in range(h):
   for j in range(w):
       img[i,j] = new_hist[img[i,j]]
       

plt.imshow(img, cmap='gray')
```

# 	Viết hàm: lọc ảnh với các bộ lọc: median, smoothing, sharpening (Median, Lena, Sharpen)
 

```python
#Lọc Median

import cv2  
import numpy as np 
import matplotlib.pyplot as plt 


# Định nghĩa hàm lọc median
def median_filter(img):
    m, n = img.shape
    img_new = np.zeros([m, n])
    for i in range(1, m-1): 
        for j in range(1, n-1): 
            temp = [img[i-1, j-1], 
                   img[i-1, j], 
                   img[i-1, j + 1], 
                   img[i, j-1], 
                   img[i, j], 
                   img[i, j + 1], 
                   img[i + 1, j-1], 
                   img[i + 1, j], 
                   img[i + 1, j + 1]] 
          
            temp = sorted(temp) 
            img_new[i, j]= temp[4] 
    img_new = img_new.astype(np.uint8)
    return img_new

fig = plt.figure(figsize=(16, 9)) 
ax1, ax2 = fig.subplots(1, 2)

# Đọc và hiển thị ảnh gốc
image = cv2.imread('Median.jpg', 0)
ax1.imshow(image, cmap='gray')
ax1.set_title("ảnh gốc")

# Lọc Median và hiển thị ảnh
imagenew = median_filter(image) 
ax2.imshow(imagenew, cmap='gray')
ax2.set_title("ảnh lọc median")
plt.show()
```

```python
# Lọc Smoothing, chọn loại bộ lọc Gaussian 3x3

import cv2  
import numpy as np 
import matplotlib.pyplot as plt 


def Tich_chap(img,mask):
    m, n = img.shape
    img_new = np.zeros([m, n])
    for i in range(1, m-1):
        for j in range(1, n-1):
            temp   =  img[i-1, j-1]    * mask[0, 0]\
                   +  img[i-1, j]      * mask[0, 1]\
                   +  img[i-1, j + 1]  * mask[0, 2]\
                   +  img[i, j-1]      * mask[1, 0]\
                   +  img[i, j]        * mask[1, 1]\
                   +  img[i, j + 1]    * mask[1, 2]\
                   +  img[i + 1, j-1]  * mask[2, 0]\
                   +  img[i + 1, j]    * mask[2, 1]\
                   +  img[i + 1, j + 1]* mask[2, 2]
            img_new[i, j]= temp
    img_new = img_new.astype(np.uint8)
    return img_new

Gaussian3x3_filter = np.array(([0.0751/4.8976, 0.1238/4.8976, 0.0751/4.8976],
                              [0.1238/4.8976, 0.2042/4.8976, 0.1238/4.8976],
                              [0.0751/4.8976, 0.1238/4.8976, 0.0751/4.8976]), dtype="float")

fig = plt.figure(figsize=(16, 9)) 
ax1, ax2 = fig.subplots(1, 2)

# Đọc và hiển thị ảnh gốc
img = cv2.imread('Lena.png', 0)
ax1.imshow(img, cmap='gray')
ax1.set_title("ảnh gốc Lena")

# Lọc Smoothing và hiển thị ảnh
imagenew = Tich_chap(img, Gaussian3x3_filter) 
ax2.imshow(imagenew, cmap='gray')
ax2.set_title("Ảnh sau khi lọc smoothing")
plt.show()
```

```python
# Lọc Shapening
Laplacian_filter = np.array(([0, 1, 0],
                               [1, -4, 1],
                               [0, 1, 0]), dtype="float")



fig = plt.figure(figsize=(16, 9)) 
ax1, ax2 = fig.subplots(1, 2)
# Đọc và hiển thị ảnh gốc
img = cv2.imread('Sharpen.png', 0)
ax1.imshow(img, cmap='gray')
ax1.set_title("ảnh gốc sharpen")

# Lọc Smoothing và hiển thị ảnh
imagenew = Tich_chap(img, Laplacian_filter) 
ax2.imshow(imagenew, cmap='gray')
ax2.set_title("Ảnh sau khi lọc sharpen")
plt.show()
```
